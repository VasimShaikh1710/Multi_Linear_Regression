```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df5.columns = list(df4)
```

```python
df5
```

```python
X = df5.drop(['Price','Age','KM'], axis = 1)
X
```

```python
Y = df5['Price']
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
LR.fit(X_train,Y_train)
Y_pred_train = LR.predict(X_train)
Y_pred_test = LR.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
R2_Train = mean_squared_error(Y_pred_train, Y_train)
R2_Test = mean_squared_error(Y_pred_test, Y_test)
R2_Train.round(2)
R2_Test.round(2)
```

```python
Training_error = np.sqrt(R2_Train)
Testing_error = np.sqrt(R2_Test)
print("Training_error",Training_error.round(2))
print("Testing_error",Testing_error.round(2))
```

```python
from sklearn.linear_model import Ridge
```

```python
RD = Ridge(alpha = 500)
RD.fit(X_train,Y_train)

Y_pred_train = RD.predict(X_train)
Y_pred_test = RD.predict(X_test)

from sklearn.metrics import mean_squared_error
R2_Train = mean_squared_error(Y_pred_train, Y_train)
R2_Test = mean_squared_error(Y_pred_test, Y_test)
print(R2_Train.round(2))
print(R2_Test.round(2))

Training_error = np.sqrt(R2_Train)
Testing_error = np.sqrt(R2_Test)
print("Training_error",Training_error.round(2))
print("Testing_error",Testing_error.round(2))
```
